# Create a word cloud from the concatenated abstract file

In the previous tutorial `Zotero_RevrieveAbstractsFromCollection`, I concatenated all of the abstracts from papers in a certain Zotero collection together. 

In this code, I load that file, and create a simple word file from the text. 

For more information on customizing the word cloud, see [WordCloud on GitHub](https://github.com/amueller/word_cloud). Code here was taken from the [Minimal Example](https://github.com/amueller/word_cloud/blob/master/examples/simple.py).

```python
from wordcloud import WordCloud
import pickle
import matplotlib.pyplot as plt

# Read the whole text.
file = open('concatenated_abstracts.txt', 'rb')
text = pickle.load(file)
print(text[0:100])

# Generate a word cloud image
wordcloud = WordCloud(width=1200,height=600,scale=1).generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
wordcloud.to_file('Primate-Hippocampal-Ephys_abstract_wordcloud.png')

# lower max_font_size
wordcloud = WordCloud(width=1200,height=600,scale=1,max_font_size=150).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file('Primate-Hippocampal-Ephys_abstract_wordcloud-max_150.png')
```
